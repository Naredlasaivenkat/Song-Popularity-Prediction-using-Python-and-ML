## Performance Evaluation

```python
a = []
b = []
c = []

def performance_metrics(algorithm, predict, testY):
    mse = mean_squared_error(testY, predict)
    mae = mean_absolute_error(testY, predict)
    r2 = r2_score(testY, predict)
    a.append(mse)
    b.append(mae)
    c.append(r2)
    print(algorithm + ' Mean Squared Error: {:.4f}'.format(mse))
    print(algorithm + ' Mean Absolute Error: {:.4f}'.format(mae))
    print(algorithm + ' R^2 Score: {:.4f}'.format(r2))

    # Plotting best-fit line
    plt.scatter(testY, predict)
    plt.plot([testY.min(), testY.max()], [testY.min(), testY.max()], '--r', lw=2)  # best-fit line
    plt.xlabel('True Values')
    plt.ylabel('Predictions')
    plt.title(algorithm + ' Predictions vs True Values')
    plt.show()
```

## xgboost Classifier

```python
import xgboost as xgb
import joblib
import os

if os.path.exists('model/XGBoostRegressor_weights.pkl'):
    # Load the model from the pkl file
    regressor = joblib.load('model/XGBoostRegressor_weights.pkl')
    predict = regressor.predict(x_test)
    performance_metrics("model/XGBoost Regressor", predict, y_test)
else:
    # Train the regressor on the training data
    regressor = xgb.XGBRegressor()
    regressor.fit(x_train, y_train)
    # Make predictions on the test data
    predict = regressor.predict(x_test)
    # Save the model weights to a pkl file
    joblib.dump(regressor, 'model/XGBoostRegressor_weights.pkl')
    print("XGBoost Regressor model trained and model weights saved.")
    performance_metrics("XGBoost Regressor", predict, y_test)
```

## CatBoost Classifier

```python
from catboost import CatBoostRegressor
import joblib
import os

catboost_model_path = 'model/CatBoost_model.pkl'

if os.path.exists(catboost_model_path):
    catboost_model = joblib.load(catboost_model_path)
else:
    catboost_model = CatBoostRegressor(iterations=1000, learning_rate=0.1, depth=6, silent=True)  # Adjust parameters as needed
    catboost_model.fit(x_train, y_train)

    joblib.dump(catboost_model, catboost_model_path)

y_pred_catboost = catboost_model.predict(x_test)

performance_metrics('CatBoost Regressor', y_pred_catboost, y_test)
```

## BayesianRidge Classifier

```python
from sklearn.linear_model import BayesianRidge
import joblib
import os

bayesian_ridge_model_path = 'model/BayesianRidge_model.pkl'

if os.path.exists(bayesian_ridge_model_path):
    bayesian_ridge_model = joblib.load(bayesian_ridge_model_path)
else:
    bayesian_ridge_model = BayesianRidge()
    bayesian_ridge_model.fit(x_train, y_train)

    joblib.dump(bayesian_ridge_model, bayesian_ridge_model_path)

y_pred_bayesian_ridge = bayesian_ridge_model.predict(x_test)

performance_metrics('Bayesian Ridge Regression', y_pred_bayesian_ridge, y_test)
```

## ElasticNet Classifier

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
model_path = 'model/ElasticNetRegressor_weights.pkl'
if os.path.exists(model_path):
    # Load the model from the pkl file
    regressor = joblib.load(model_path)
    predict = regressor.predict(x_test)
    performance_metrics("ElasticNet Regressor", predict, y_test)
else:
    # Train the regressor on the training data
    regressor = ElasticNet()
    regressor.fit(x_train, y_train)
    # Make predictions on the test data
    predict = regressor.predict(x_test)
    # Save the model weights to a pkl file
    joblib.dump(regressor, model_path)
    print("ElasticNet Regressor model trained and model weights saved.")
    performance_metrics("ElasticNet Regressor", predict, y_test)
```
